```python
#concatenate features and target into a single dataframe
#axis = 1 makes it concatenate column wise
df = pd.concat([features, target], axis = 1)
df
```

```python
#use round(decimals = 2) to set precision upto 2 decimal places
df.describe().round(decimals = 2)
```

```python
# calc correlation between every column in the data
corr = df.corr('pearson')

# take absolute correlation value between each attribute and the target
corrs = [abs(corr[attr]['target']) for attr in list(features)]

# make a list of pairs [(corr, feature)]
l = list(zip(corrs, list(features)))

# sort the list of pairs in descending order
# with correlation value as the key for sorting
l.sort(key = lambda x : x[0], reverse = True)

# "unzip" pairs to two lists
# zip(*l) - takes a list that looks like [[a,b,c], [d,e,f], [g,h,i]]
# and returns [[a,d,g], [b,e,h], [c,f,i]]
corrs, labels = list(zip((*l)))

# plot correlations as a wrt target variable as a bar graph
index = np.arange(len(labels))
plt.figure(figsize = (15,5))
plt.bar(index, corrs, width = 0.5)
plt.xlabel('Attributes')
plt.ylabel('Correlation with the target variable')
plt.xticks(index, labels)
plt.show()
```

NORMALIZE THE DATA

```python
X = df['LSTAT'].values
Y = df['target'].values
```

```python
# before normalisation
print(Y[:5])
```

```python
x_scaler = MinMaxScaler()
X = x_scaler.fit_transform(X.reshape(-1,1))
X = X[:, -1]

y_scaler = MinMaxScaler()
Y = y_scaler.fit_transform(Y.reshape(-1,1))
Y = Y[:, -1]
```

```python
# after normalisation
print(Y[:5])
```

Step 2: Define Error

Mean Square Error



```python
def error(m, x, c, t):
    N = x.size
    e = sum(((m * x + c) - t) ** 2)
    return e * 1/(2 * N)
```

Step 3: Splitting of the data

```python
# 0.2 indicates that a random 20% of the sample is split as the testing data
xtrain, xtest, ytrain, ytest = train_test_split(X, Y, test_size = 0.2)
```

Update Function

```python
def update(m, x, c, t, learning_rate):
    grad_m = sum(2 * ((m * x + c) - t) * x)
    grad_c = sum(2 * ((m * x + c) - t))
    m = m - grad_m * learning_rate
    c = c - grad_c * learning_rate
    return m, c
```

Gradient Descent: If the error falls below the error threshold, the G.D process is stopped and the weights are returned

```python
def gradient_descent(init_m, init_c, x, t, learning_rate, iterations, error_threshold):
    m = init_m
    c = init_c
    error_values = list()
    mc_values = list()
    
    for i in range(iterations):
        e = error(m, x, c, t)
        if(e < error_threshold):
            print("Stopping Gradient Descent!")
            break
        error_values.append(e)
        m, c = update(m, x, c, t, learning_rate)
        mc_values.append((m, c))
        
    return m, c, error_values, mc_values
```

```python
%%time
init_m = 0.9
init_c = 0
learning_rate = 0.001
iterations = 250
error_threshold = 0.001

m, c, error_values, mc_values = gradient_descent(init_m, init_c, xtrain, ytrain, learning_rate, iterations, error_threshold)
```

```python
# as the number of iterations increases, the changes will be less noticeable
# inorder to reduce processing time, choose small intervals
mc_values_anim = mc_values[0:250:5]
```

```python
fig, ax = plt.subplots()
ln, = plt.plot([], [], 'ro-', animated = True)


def init():
    plt.scatter(xtest, ytest, color = 'g')
    ax.set_xlim(0, 1.0)
    ax.set_ylim(0, 1.0)
    return ln,

def update_frame(frame):
    m, c = mc_values_anim[frame]
    x1, y1 = -0.5, m * -0.5 + c
    x2, y2 = 1.5, m * 1.5 + c
    ln.set_data([x1, x2], [y1, y2])
    return ln,

anim = FuncAnimation(fig, update_frame, frames = range(len(mc_values_anim)), init_func = init, blit = True)
HTML(anim.to_html5_video())
```

```python
plt.scatter(xtrain, ytrain, color = 'b')
plt.plot(xtrain, xtrain * m + c, color = 'r')
```

Plotting error values

```python
plt.plot(np.arange(len(error_values)), error_values)
plt.ylabel('Error')
plt.xlabel('Iterations')
```

Step 5: PREDICTON

```python
# calculate the predictions on the test set as a vectorized operation
predicted = (m * xtest) + c
```

```python
# compute MSE for the predicted values on the testing dataset
mean_squared_error(ytest, predicted)
```

```python
# put xtest, ytest and predicted values into a single DataFrame so that we can
# see predicted values alongside the testing dataset
p = pd.DataFrame(list(zip(xtest, ytest, predicted)), columns = ['x', 'target_y', 'predicted'])
p.head()
```
